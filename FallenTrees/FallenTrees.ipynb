Final analytics + CSV generation

```python
from ultralytics import YOLO
import cv2
import os
import csv

# Load trained model
model = YOLO("/kaggle/working/best.pt")

# Dataset test images
test_dir = "/kaggle/working/Fallen-trees-(with-palms)-1/test/images"

# Output folder
out_dir = "/kaggle/working/output_frames"
os.makedirs(out_dir, exist_ok=True)

csv_rows = []

for img_name in os.listdir(test_dir):

    if not img_name.lower().endswith((".jpg", ".jpeg", ".png")):
        continue

    img_path = os.path.join(test_dir, img_name)

    result = model(img_path, conf=0.35)[0]

    # Save annotated image
    annotated = result.plot()
    cv2.imwrite(os.path.join(out_dir, img_name), annotated)

    frame_h, frame_w = result.orig_shape

    for box in result.boxes:

        x1, y1, x2, y2 = box.xyxy[0].tolist()

        conf = float(box.conf[0])

        width = x2 - x1
        height = y2 - y1

        bbox_area = width * height
        frame_area = frame_w * frame_h

        occupancy = (bbox_area / frame_area) * 100

        # Obstruction Category
        if occupancy < 5:
            category = "Minor Obstruction"
        elif occupancy < 15:
            category = "Partial Road Block"
        elif occupancy < 30:
            category = "Major Road Block"
        else:
            category = "Critical Road Block"

        csv_rows.append({
            "image_name": img_name,
            "class": "fallen_tree",
            "confidence": round(conf, 3),
            "x": round(x1),
            "y": round(y1),
            "width": round(width),
            "height": round(height),
            "occupancy_percent": round(occupancy, 2),
            "obstruction_category": category
        })

# Save CSV
csv_path = "/kaggle/working/bounding_box_results.csv"

with open(csv_path, "w", newline="") as f:
    writer = csv.DictWriter(
        f,
        fieldnames=csv_rows[0].keys()
    )
    writer.writeheader()
    writer.writerows(csv_rows)

print(f"Annotated images saved to: {out_dir}")
print(f"CSV saved to: {csv_path}")
print(f"Total detections: {len(csv_rows)}")
```
